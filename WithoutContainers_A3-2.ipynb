```python
#Scaling X Variables

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_rescaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print(df_rescaled.head(5))

print(df_rescaled.describe())
```

```python
#Correlation Matrix

data_clean = df_rescaled.copy()
data_clean["y"] = y


corr_matrix = data_clean.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
print(df_rescaled.shape)
print(y.shape)
```

```python
#Splitting Data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_rescaled, y, test_size=0.25, random_state=7)

print('Shape of training data :',X_train.shape)
print('Shape of testing data :',y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model = LinearRegression()

model.fit(X_train, y_train)

predict_train = model.predict(X_train)

r2_train = r2_score(y_train, predict_train)
print('\nR^2 score on train dataset: ', r2_train)

mae_train = mean_absolute_error(y_train, predict_train)
print('\nMean Absolute Error on train dataset: ', mae_train)

mse_train = mean_squared_error(y_train, predict_train)
print('\nMean Squared Error on train dataset: ', mse_train)

predict_test = model.predict(X_test)

r2_test = r2_score(y_test, predict_test)
print('\nR^2 score on test dataset: ', r2_test)

mae_test = mean_absolute_error(y_test, predict_test)
print('\nMean Absolute Error on test dataset: ', mae_test)

mse_test = mean_squared_error(y_test, predict_test)
print('\nMean Squared Error on test dataset: ', mse_test)
```

```python
import statsmodels.api as sm

X_train_with_intercept = sm.add_constant(X_train)

model_stats = sm.OLS(y_train, X_train_with_intercept).fit()

print(model_stats.summary())

print("\nConfidence intervals for coefficients:")
print(model_stats.conf_int())
```
